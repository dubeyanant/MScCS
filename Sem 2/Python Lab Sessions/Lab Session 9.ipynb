```python
#Q.1 Write a program to implement inheritance, create a class named as polygon.
#Input the number of sides of a polygon and display them.
#Create a class named as triangle which inherits properties from polygon and calculates and displays area of the triangle.

class Polygon:
    def __init__(self, num_sides):
        self.num_sides = num_sides

    def display_sides(self):
        print(f"This polygon has {self.num_sides} sides.")

class Triangle(Polygon):
    def __init__(self, a, b, c):
        super().__init__(3)  # Initialize Polygon with 3 sides
        self.a = a
        self.b = b
        self.c = c

    def calculate_area(self):
        # Using Heron's formula to calculate the area
        s = (self.a + self.b + self.c) / 2
        area = (s * (s - self.a) * (s - self.b) * (s - self.c)) ** 0.5
        return area

    def display_area(self):
        area = self.calculate_area()
        print(f"The area of this triangle is {area}.")


t = Triangle(3, 4, 6)
t.display_sides()  # This polygon has 3 sides.
t.display_area()   # The area of this triangle is 6.0.
```

```python
#Q.2 Write a program to create base class Vehicle and derived classes Bus and Car.
#Implement inheritance, include required attributes and methods for all the three classes.

class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def start(self):
        print("Vehicle started")

    def stop(self):
        print("Vehicle stopped")

class Bus(Vehicle):
    def __init__(self, make, model, year, seating_capacity):
        super().__init__(make, model, year)
        self.seating_capacity = seating_capacity

    def open_doors(self):
        print("Bus doors opened")

    def close_doors(self):
        print("Bus doors closed")

class Car(Vehicle):
    def __init__(self, make, model, year, num_wheels):
        super().__init__(make, model, year)
        self.num_wheels = num_wheels

    def change_gear(self):
        print("Car gear changed")

    def turn(self, direction):
        print(f"Car turned {direction}")

        
        
bus = Bus("Volvo", "B7L", 2005, 50)
car = Car("Toyota", "Corolla", 2010, 4)

print(bus.make, bus.model, bus.year, bus.seating_capacity)
bus.start()
bus.open_doors()

print(car.make, car.model, car.year, car.num_wheels)
car.start()
car.change_gear()
car.turn("left")
```

```python
#Q.3 Write a Python class to find the three elements that sum to zero from a set of n real numbers.

class FindTriplets:
    def __init__(self, numbers):
        self.numbers = numbers

    def find_triplets(self):
        n = len(self.numbers)
        result = []
        self.numbers.sort()

        for i in range(n - 2):
            left = i + 1
            right = n - 1

            while left < right:
                total = self.numbers[i] + self.numbers[left] + self.numbers[right]
                if total == 0:
                    result.append([self.numbers[i], self.numbers[left], self.numbers[right]])
                    left += 1
                    right -= 1
                elif total < 0:
                    left += 1
                else:
                    right -= 1

        return result
    
numbers = [-25, -10, -7, -3, 2, 4, 8, 10]
finder = FindTriplets(numbers)
result = finder.find_triplets()
print(result)
```

```python
#Q.4 Using class implement Queue and all its operations.

class Queue:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def enqueue(self, item):
        self.items.append(item)

    def dequeue(self):
        if self.is_empty():
            return None
        else:
            return self.items.pop(0)

    def size(self):
        return len(self.items)

    def peek(self):
        if self.is_empty():
            return None
        else:
            return self.items[0]

        
q = Queue()

print(q.is_empty())  # True
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
print(q.size())  # 3
print(q.peek())  # 1
print(q.dequeue())  # 1
print(q.size())  # 2
print(q.is_empty())  # False
```

```python
#Q.5 Using class implement Stack and all its operations.

class Stack:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if self.is_empty():
            return None
        else:
            return self.items.pop()

    def size(self):
        return len(self.items)

    def peek(self):
        if self.is_empty():
            return None
        else:
            return self.items[-1]

        
s = Stack()

print(s.is_empty())  # True
s.push(1)
s.push(2)
s.push(3)
print(s.size())  # 3
print(s.peek())  # 3
print(s.pop())  # 3
print(s.size())  # 2
print(s.is_empty())  # False
```

```python
#Q.6 Implement Breadth first Search.

from collections import defaultdict

class Graph:
    def __init__(self):
        self.graph = defaultdict(list)

    def add_edge(self, u, v):
        self.graph[u].append(v)

    def bfs(self, start):
        visited = [False] * len(self.graph)
        queue = []
        queue.append(start)
        visited[start] = True

        while queue:
            node = queue.pop(0)
            print(node, end=' ')

            for neighbor in self.graph[node]:
                if not visited[neighbor]:
                    queue.append(neighbor)
                    visited[neighbor] = True

                    
g = Graph()

# add edges to graph
g.add_edge(0, 1)
g.add_edge(0, 2)
g.add_edge(1, 2)
g.add_edge(2, 0)
g.add_edge(2, 3)
g.add_edge(3, 3)

# perform BFS starting from node 2
print("BFS traversal starting from node 2:")
g.bfs(2)
```

```python
#Q.7 Implement Depth First Search.

from collections import defaultdict

class Graph:
    def __init__(self):
        self.graph = defaultdict(list)

    def add_edge(self, u, v):
        self.graph[u].append(v)

    def dfs_util(self, node, visited):
        visited[node] = True
        print(node, end=' ')

        for neighbor in self.graph[node]:
            if not visited[neighbor]:
                self.dfs_util(neighbor, visited)

    def dfs(self, start):
        visited = [False] * len(self.graph)
        self.dfs_util(start, visited)

        
g = Graph()

# add edges to graph
g.add_edge(0, 1)
g.add_edge(0, 2)
g.add_edge(1, 2)
g.add_edge(2, 0)
g.add_edge(2, 3)
g.add_edge(3, 3)

# perform DFS starting from node 2
print("DFS traversal starting from node 2:")
g.dfs(2)
```
